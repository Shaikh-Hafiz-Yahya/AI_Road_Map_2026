### difference b/w 
### append() , extend()

```python
# append()
l1 , l2 = [1,2,3,4,5] , [10,20,30,40,50]
empty_lst = []

empty_lst.append(l1)
print(empty_lst)
print(empty_lst[0])

empty_lst.append(l2)
print(empty_lst)
print(empty_lst[1])
```

```python
# extend()
l = [100 , 1100 ,22000 ,330000]
l.extend([4400000 , 5500000 , 6600000])
print(l)
print(l[:-3])
```

```python
# update()

# Multiple elements add karta hai ye three form sy(list/tuple/set)
s = {1 , 2}
s.update([4 , 6 , 8])
print(s)
# similarly
s = {1, 2}
s.update((3, 4, 5))
print(s)
# simiarly
s = {1, 2}
s.update({3, 4, 5})
print(s)
```

```python
# Ek set me fruits ke 3 naam update() se add karo.
fruits = {'Orange' , 'Banana' , 'Applae' , 'Kiwi'}
fruits.update(['Pineapple' , 'Guava' , 'Watermelon'])
fruits
```

```python
# remove()
#  Element remove karta hai
#  Agar element na ho → error
s = {1000 , 10000 , 10 , 50000}
s.remove(10)
s
```

```python
# Practice:
# Set se ek number remove karo jo maujood ho.
s = {11 , 101 , 10 , 1 , 201}
s.remove(1)
s
```

```python
# discard()
# Element remove karta hai
# Agar na ho → error nahi
s =  {10.1 , 5.2 , 30.99 , 1.1}
s.discard(1.1)
print(s)

s2 = {3.1 , 2.22 , 99.9 , 10.11 , 20.11} # Agar ele na ho → error nah milay gai
s2.discard(77.7)
print(s2)

```

```python
# difference b/w 
# remove() , discard()
s1 = {111 , 1 , 221 , 0 , 90 , 22}
s1.remove(333)  #error --> Reason '333' set main moujood hi nahi
print(s1)
```

```python
# discard()
s2 = {111 , 1 , 221 , 0 , 90 , 22}
s2.discard(333) #yahan error nahi milay ga --># Agar element na ho set main → error nah milay gai
print(s2)
```

```python
# pop()
# Random element remove + return karta hai
s = {1 , 3 , 0 , 4}
x = s.pop()
print(x)
print(s)

```

```python
# union()
# Do sets ko combine karta hai
SubFieldsOfAI = {
    'Machine Learning',
    'Deep Learning',
    'ANN',
    'Generative AI'
}

DataTypes = {
    'Structured Data (DB)',
    'Unstructured Data (Text, PDF, Audio, Video(Sequence_of_images), Images)',
    'Vectors',
    'Matrices'
}

com = SubFieldsOfAI.union(DataTypes)
print(com)
```
