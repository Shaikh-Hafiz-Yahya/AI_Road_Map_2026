```python
class Employee:
    
    #static var / class 
    HeadOffice = 'XYZ'
    Time = 8

    def __init__(self , name , age , salary):

        # instance var / Attr=var 
        self.e_name = name
        self.e_age = age
        self.e_salary = salary

    # instance method
    def emp_info(self):
        print(f'Employee:{self.e_name}\nAge:{self.e_age}\nPosition:{self.e_salary}')

    # class method
    @classmethod
    def head_office(cls):
        print(f'Head Office IS In:{cls.HeadOffice}')

empA:Employee = Employee('John' , 28 , 200000)
empB:Employee = Employee('Emily' , 23 , 350000)

empB.emp_info()

Employee.head_office() #best way class method call-->through the class name but required '@classmethod'
```

```python
Holiday = 2
class Employee:
    
    #static var / class 
    HeadOffice = 'XYZ'
    Time = 8

    def __init__(self , name , age , salary):

        # instance var / Attr=var 
        self.e_name = name
        self.e_age = age
        self.e_salary = salary

    # instance method
    def emp_info(self):
        print(f'Employee:{self.e_name}\nAge:{self.e_age}\nPosition:{self.e_salary}')

    # class method
    @classmethod
    def head_office(cls):
        print(f'Head Office IS In:{cls.HeadOffice}')

    #static method
    @staticmethod
    def head_office_holiday():
        print(f'HeadOffice Holidays per week = {Holiday}')    

empA:Employee = Employee('John' , 28 , 200000)
empB:Employee = Employee('Emily' , 23 , 350000)

empB.emp_info()

Employee.head_office() #best way class method call-->through the class name but required '@classmethod'

Employee.head_office_holiday() #best way Static method call-->through the class name but required '@staticmethod'
```
