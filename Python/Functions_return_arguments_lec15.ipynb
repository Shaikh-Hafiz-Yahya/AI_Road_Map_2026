```python
def df_dz(x , y):
    
    dfdx = 2 * x
    dfdy = 3 * (y ** 2)

    dzdx = 1
    dzdy = 1

    dfdz = dfdx * dzdx + dfdy * dzdy
    return dfdz


print(df_dz(1, 4))
```

```python
def derivative(x , *y):

    for i in y:
        # Q: f(z) = (x**2 + y**3)
        dz_over_dx = 2*(x)**(2-1) * (1) 
        dz_over_di = 3*(i)**(3-1) * (1)
        differentiate_w_r_to_z_is = (dz_over_dx + dz_over_di)
        
        print(f"dx/dz = {dz_over_dx}")
        print(f"dy/dz = {dz_over_di}")
        print("f(z) = d(x,y)/dz = " , differentiate_w_r_to_z_is)  

derivative(1 , 4 , 3 , 5 , 9)
```
