# Inner Class

```python
class Employee:

    # parameterized constructor
    def __init__(self , e_name , e_position):

        # Attr = var 
        self.name = e_name
        self.position = e_position
    
    # instance method 
    def display_info(self):
        print(f'Employee:{self.name}\nAge:{self.position}')

    class Address:
        
        def __init__(self , emp_1 , emp_2 , emp_n):

            self.emp_1 = emp_1
            self.emp_2 = emp_2
            self.emp_n = emp_n

        def diplay_address(self):
            print(f'Employee1:{self.emp_1}\nEmployee:{self.emp_2}\nEmployee:{self.emp_n}')   

emp1 = Employee('John' , 'Manager')
emp2 = Employee('JefferyHinton' , 'Supervisor')

emp1.Address = Employee.Address('street N001 Canada' , 'street L015 Canada' , 'street QK190K Canada')
emp2.Address = Employee.Address('street Z01G Canada' , 'street Ao015 Canada' , 'street Qp110K Canada')

print('Call Inner Class Method:')
emp1.Address.diplay_address()

print('\nCall Outer Class Method:')
emp2.display_info()
```

```python
class Animal:
    
    #Parameterized Constructor
    def __init__(self , name , color):
        
        #Attr = Var / instance var
        self.b_name = name
        self.b_color = color

    def Wild(self):
        print(f"This is a {self.b_name}\nColor is {self.b_color}")

    # inner class 
    class WildLife:
        
        #Parameterized Constructor
        def __init__(self , hunt , advanture):
            
            #Attr = Var / instance var
            self.lion = hunt
            self.expert = advanture

        #Parameterized Constructor
        def Forest(self):
            print(f"Welcome to {self.lion}\nWild_Life_Lover {self.expert}")
# objects
beauti1 = Animal('Panda' , 'black_&_white')
beauti2 = Animal('Panther' , 'Black')

beauti1.WildLife = Animal.WildLife('Deer' , 'Discover_Animal_Life')
beauti2.WildLife = Animal.WildLife('Zebra' , 'Danger_Zone')

# call inner class method
beauti1.WildLife.Forest()

beauti2.WildLife.Forest()

# # call outer class method
beauti1.Wild()

beauti2.Wild()
```
