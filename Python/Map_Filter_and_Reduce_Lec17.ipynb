## filter function 

```python
num_list = [1 , 2 , 3 , 4 , 5]

even = list(filter(lambda n : n%2 == 0 , num_list))
print(f'list only even numbers = {even}')
```

```python
numeric_tuple = (1 , 3 , 5 , 7 , 8 , 9)

odd = tuple(filter(lambda n : n%2 == 1 , numeric_tuple))
odd
```

## ruduce function

```python
# .syntax reduce() --> reduce(lambda total , y : total + y , iteration) 
# .lambda main 2 hi argument pss hogi jb jb lambda fn reduce() k saath use hoga

from functools import reduce
NumList = [1 , 2 , 3 , 4 , 5]

redu = reduce(lambda total , y : total + y , NumList)
print(redu)

NumericList = [1 , 2 , 3 , 4 , 5]

r = reduce(lambda n , m : n+m , NumericList)
print(r)
```

```python
num_list = [1 , 3 , 5]

re = reduce(lambda m , x : m * x , num_list)
re
```

```python
n_list = [1 , 2 , 4 , 6]

# Working Rule reduce function:
# total = 1 + next = 2 = 3
# total = 3 + next = 4 = 7
# total = 7 + next = 6 = 13
# output = 13        

re = reduce(lambda total , next : total + next  , n_list)
print(f'Output = {re}')
```
