### Identity Operators

is , is not

```python
x = [1 , 2 , 4 , 9]
y = x
z = [10 , 100 , 10000]

# is 
print(y is x)
print(y is z)
print(z is z)

# is not 
print(x is not y)
print(z is not y)
```
