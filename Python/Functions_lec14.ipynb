```python
sales_rate = 150
txt_rate = 0.05
total_amount = sales_rate * txt_rate
print(total_amount)

# now above code covert user defined function
# better way.
def tax_calculator(sales_rate):

    total_amount = sales_rate * txt_rate
    print(total_amount)

txt_rate = 0.05
tax_calculator(150)    
tax_calculator(110)
tax_calculator(8100)
```

```python
def tax_calculator(sales_rate , txt_rate):

    total_amount = sales_rate * txt_rate
    print(total_amount)

tax_calculator(150 , 0.05)
tax_calculator(110 , 0.05)
tax_calculator(8100 , 0.05)
```
