```python
for wishes in range(10 , 30): #set range(10 to n-1)
    print(f"BirthDay:{wishes}")
```

```python
for wishes in range(1 , 10):
    if wishes == 5:
        continue
    else:
        print(f'Birthday:{wishes}')
```

```python
for wishes in range(1 , 12 , 2):
    if wishes == 7:
        continue
    else:
        print(f'BirthDay:{wishes}')
```

```python
for wishes in range(1 , 12):
    if wishes == 5:
        break
    else:
        print(f'BirthDay:{wishes}')
```

```python
for wishes in range(1 , 12 , 2):
    if wishes == 3 or wishes == 9:
        break
    else:
        print(f'Birthday:{wishes}')
```

Difference B/w Above & Below

```python
for wishes in range(1 , 12 , 2):
    if wishes == 3 and wishes == 9:
        break
    else:
        print(f'BirthDay:{wishes}')
```

```python
for wishes in range(1 , 12 , 3):
    if wishes == 7:
        continue
    else:
        print(f'BirthDay:{wishes}')
print('Loop is ended')        
```
