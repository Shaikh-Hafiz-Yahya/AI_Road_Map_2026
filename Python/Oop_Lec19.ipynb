# OBJECT ORIENTED PROGRAMMING

```python
# CREATE CLASS 
class Employee:
    
    #Non parameterized constructor 
    def __init__(self):

        #Attributes
        self.emp_depart = 'AI'
        self.emp_salary = 200000

#Create Object 
john = Employee()
mike_hertz = Employee()

print(mike_hertz)
```

```python
# CREATE CLASS 
class Employee:
    #Non parameterized constructor 
    def __init__(self):

        #Attributes
        self.emp_depart = 'AI'
        self.emp_salary = 200000

#Create Object 
john = Employee()
mike_hertz = Employee()

print(mike_hertz.emp_salary) #call attributes
```

```python
# CREATE CLASS 
class Employee:
    
    #Non parameterized constructor 
    def __init__(self): 

        #Attributes
        self.EmployeeDepartA , self.EmployeeDepartB  = 'AI' , 'Business' 
        self.SalaryEmployeeA , self.SalaryEmployeeB = 200000 , 90000

#Create Object 
john = Employee()
mike_hertz = Employee()

# CALL OBJ
print(john.EmployeeDepartA) 
print(mike_hertz.SalaryEmployeeB)
```
