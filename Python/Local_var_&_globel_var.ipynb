### global variable --> jo fn k insight or outsight call kiya jata ha 

```python
# local var 
def hello2():

    loc_var = 'hello_py' ## local var 
    print(loc_var) #call insight fn

hello2()    
```

```python
# local var 
def hello2():

    loc_var = 'hello_py' # local var 
    print(loc_var) #var call insight the fn

hello2()   
print(loc_var) #error (Reason var call outside the fn)
```

```python
# global var 

glo_var = 'hello_py'
def hello2():

    print(glo_var) #glo_var call insight the fn
    print(glo_var) #glo_var call insight the fn
 
hello2()   

print(glo_var) #glo_var call outsight the fn
print(glo_var)
```

## Anonymous / Lambda fn

```python
# normal functions
def add(x):

    results = x+10
    print(results)

add(3)   


# Anonymous / lambda function 

addition = lambda x:x+10 # ---> lambda ki power/taqat

result = addition(3)
print(result)
```

```python
y = lambda m,x,b : m*x + b 

eq_of_straight_line = y(50 , 4 , 100)
eq_of_straight_line
```

```python
# logic building Question: ∫(x.sinx)dx
# Apply any logic but answer lazmi (True) ho 
# formula:I.BY.P.M = ∫(u.v)dx = u.∫vdx - ∫(d/dx(u).∫vdx)dx
# consider: sinx = 3 , x = 10


integ_by_part_method = lambda u , v ,  dx : u*v*dx - ((0)*u*v*dx) * dx  
u_dot_v_dx = integ_by_part_method(10 , 3  , 1)

integration = str(u_dot_v_dx) + 'x'

question_integration , ans_integration = '∫(10*3)dx' , integration   # Let, sinx = 3 , x = 10
print(f"{question_integration} = {ans_integration}")
```
