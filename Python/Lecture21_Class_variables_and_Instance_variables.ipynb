# Class variables and Instance variables

```python
class Employee:

    #Class Variable / Static Variable
    Salary = 180000
    Depart = 'Business'

    #Parameterized Constructor
    def __init__(self , name , age , position):

        #Attributes=Variables / (instance Variable/Dynamic Variable)
        #Instance var ka connection class object sy hota ha 
        self.e_name = name
        self.e_age = age
        self.e_position = position
    
    #Behaviour / Method
    def display_info(self):
        print(f'Employee:{self.e_name}\nAge:{self.e_age}\nPosition:{self.e_position}')

emp1:Employee = Employee('John' , 29 , 'Manager')
emp2:Employee = Employee('JefferyHinton' , 26 , 'Supervisor')

emp1.display_info()
emp2.display_info()
```
