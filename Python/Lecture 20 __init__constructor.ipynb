# parameterized constructor

```python
class Employee:

    #parameterized constructor
    def __init__(self , emp_depart , emp_salary):
        
       self.emp_depart = emp_depart
       self.emp_salary = emp_salary

john = Employee('AI' , 200000) 
mike_hertz = Employee('Business' , 90000)

print(f"Salary is {john.emp_salary}")
print(f"Depart {mike_hertz.emp_depart}")
```

```python
class Employee:

    #parameterized constructor
    def __init__(self , emp_depart , emp_salary):
        
       self.EmployeeDepart = emp_depart
       self.EmployeeSalary = emp_salary

john = Employee('AI' , 200000) 
mike_hertz = Employee('Business' , 90000)

print(f"Salary is {john.EmployeeSalary}")
print(f"Depart {mike_hertz.EmployeeDepart}")
```

```python
class Employee:

    def __init__(self , name , position , salary):
        
        #Attributes/var
        self.name = name
        self.position = position
        self.salary = salary
    # Behaviour/Method
    def display(self):
        print(f'EmployeeName:{self.name}\nEmployeePosition:{self.position}\nEmployeeSalary = {self.salary}') 

# instatiation --> kici class ko object k through call krein ye method ha instatiation
emp1 = Employee('MacHinton' , 'Manager' , 290000)
emp2 = Employee('John' , 'DataScientist' , 500000)

emp1.display()
print('--------------')
emp2.display()
```

```python
class Employee:

    def __init__(self , name , position , salary):
        
        #Attributes/var
        self.name = name
        self.position = position
        self.salary = salary

    # Behaviour/Method
    def display(self):
        print(f'EmployeeName:{self.name}\nEmployeePosition:{self.position}\nEmployeeSalary = {self.salary}') 

#TypeHunting
emp1:Employee = Employee('MacHinton' , 'Manager' , 290000)
emp2:Employee = Employee('John' , 'DataScientist' , 500000)

# Call obj
emp1.display()
print('..................')
emp2.display()
```
