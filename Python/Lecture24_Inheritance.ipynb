# INHERITANCE

```python
class A:

    def method1(self):
        print('This is method1')

    
    def method2(self):
        print('This is method2')

class B(A):

    
    def method3(self):
        print('This is method3')

        
    def method4(self):
        print('This is method4')    

class C(A):

    
    def method5(self):
        print('This is method5')
        
    def method6(self):
        print('This is method6')          

a:A = A()
a.method1()

b:B = B()
b.method1()
b.method3()

c:C = C()

c.method1()
c.method5()
```

## Multilevel Inheritance
ðŸ“Œ Definition:

Jab ek class dusri class se inherit kare, aur phir teesri class us second class se inherit kare â€” is chain ko multilevel inheritance kehte hain.

ðŸ‘‰ Matlab: Grandparent â†’ Parent â†’ Child

```python
class A:

    def method1(self):
        print('This is method1')

    
    def method2(self):
        print('This is method2')

class B(A):

    def method3(self):
        print('This is method3')

        
    def method4(self):
        print('This is method4')    

class C(B):
        
        def method5(self):
            print('This is method5')
             
        def method6(self):
            print('This is method6')     

c:C = C()

c.method1()

c.method3()

c.method5()
```

# Multiple Inheritance
ðŸ“Œ Definition:

Jab ek class ek se zyada classes se inherit kare â€” isay multiple inheritance kehte hain.

ðŸ‘‰ Matlab: Ek child, multiple parents

```python
class A:

    def method1(self):
        print('This is method1')

    
    def method2(self):
        print('This is method2')

class B:

    def method3(self):
        print('This is method3')

        
    def method4(self):
        print('This is method4')    

class C(A , B):
        
        def method5(self):
            print('This is method5')

b:B = B()
b.method3()

c:C = C()

c.method2()
```
