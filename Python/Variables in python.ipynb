## Concatenation 
there are 3 types of concatenation 
1. plus operator(+)
2. f str method
3. dotformat method( .format )

```python
# 1. (+)
name = 'Hafiz' + 'Yahya'
print(name)
```

```python
name = 'python' + 7 #error
name
```

```python
num_str = '7'+'7' 
num_str
```

```python
# 2. f str
f_name , l_name = 'Hafiz' , 'Yahya'
print(f'{f_name}{l_name}')
# diff b/w 
print(f'{f_name} {l_name}')
```

```python
# 3. dot format 
```

```python
first_name , last_name = 'Hafiz' , 'Yahya'
full_name = first_name,last_name
print(type(full_name))
```

```python
first_name = 'Hafiz'
last_name = 'Yahya'
full_name = first_name,last_name  #strings convert to tuple
print(full_name , type(full_name))

print('\n---Difference_b/w_Above_&_Below---\n')

FullName = ('Hafiz' , 'Yahya') #This is a tuple
print(FullName , type(FullName)) 

print('\n---Difference_b/w_Above_&_Below---\n')

first_name = 'Hafiz'
last_name = 'Yahya'
full_name = first_name+last_name  #concatenation 
print(full_name , type(full_name)) 
```
