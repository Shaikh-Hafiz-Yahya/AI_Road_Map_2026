```python
#ClassB Ka constructor call nahi hoga sirf classA&C KA CALL hoga (Reason Left to right rule mtlb
#k A pehly likha ha inheritance main YAHI CONCEPT METHOD RESOLUTION ORDER KA HA(MRO)) 
   
class A:

    def __init__(self):
        print('This constructor is classA')

    def method1(self):
        print('This is method1')

    def method2(self):
        print('This is method2')

class B:

    def __init__(self):
        print('This constructor is classB')

    def method3(self):
        print('This is method3')

    def method4(self):
        print('This is method4') 

class C(A , B): #Left to right rule
    
    def __init__(self):

        super().__init__()
        print('This constructor is classC')

    def method5(self):
        print('This is method5')

    def method6(self):
        print('This is method6') 

c:C = C()  
c 
   
```

# Important Concept (Advanced but Useful)

Multiple inheritance mein kabhi kabhi method conflict hota hai (agar dono parent classes mein same method ho).

Python isko solve karta hai using:

ðŸ‘‰ MRO (Method Resolution Order)

```python
#ClassB Ka constructor call nahi hoga sirf classA&C KA CALL hoga (Reason Left to right rule mtlb
#k A pehly likha ha inheritance main YAHI CONCEPT METHOD RESOLUTION ORDER KA HA(MRO)) 
   
class A:

    def __init__(self):
        print('This constructor is classA')

    def method1(self):
        print('This is method1')

    def method2(self):
        print('This is method2')

class B:

    def __init__(self):
        print('This constructor is classB')

    def method1(self):
        print('This is method3')

    def method2(self):
        print('This is method4') 

class C(A , B): #Left to right rule
    
    def __init__(self):

        super().__init__()
        print('This constructor is classC')

    def method3(self):
        print('This is method5')

    def method4(self):
        print('This is method6') 

c:C = C()  

c.method1()
c.method2()
```
