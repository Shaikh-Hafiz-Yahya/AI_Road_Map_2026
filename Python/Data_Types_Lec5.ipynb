### sequence , list , tuple , range

```python

lst = [10 , 20 , 30 , 40 , 50]
type(lst)
```

```python
tup = (1 , 2 , 3 , 4 , 5)
type(tup)
```

```python
r = range(18)
print(r , type(r))

# convert list 
lst = list(r)
print(lst)
print(type(lst))

# convert tuple 
tup = tuple(r)
print(tup)
print(type(tup))
```
